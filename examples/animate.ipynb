```python
import numpy as np
import cloudglancer as cg
from IPython.display import Image
```

```python
# Generate a swirly point cloud so the rotation is visible
n = 1000
t = np.linspace(0, 6 * np.pi, n)
points = np.column_stack([np.cos(t) * t / 10, np.sin(t) * t / 10, t / 5])
points += 0.05 * np.random.randn(*points.shape)

fig = cg.plot(points, title="Swirl", size=2.0)


axis_style = dict(
    backgroundcolor="white",   # the colored "wall" behind each axis
    gridcolor="lightgray",     # grid lines
    showbackground=True,
    showticklabels=False,      # hide the numbers
    title="",                  # hide the x/y/z label
    zerolinecolor="lightgray",
)

fig.update_layout(
    paper_bgcolor="white",     # area outside the 3D scene
    scene=dict(
        bgcolor="white",       # area inside the 3D box
        xaxis=axis_style,
        yaxis=axis_style,
        zaxis=axis_style,
    ),
)
fig.show()
```

```python
# Render a turntable GIF rotating around the z-axis
path = cg.animate(fig, "rotation.gif", axis="z", n_frames=100, duration=60)
Image(filename=path)
```
