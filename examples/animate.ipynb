```python
import numpy as np
import cloudglancer as cg
from IPython.display import Image
```

```python
# Generate a swirly point cloud so the rotation is visible
n = 1000
t = np.linspace(0, 6 * np.pi, n)
points = np.column_stack([np.cos(t) * t / 10, np.sin(t) * t / 10, t / 5])
points += 0.05 * np.random.randn(*points.shape)

fig = cg.plot(points, title="Swirl", size=2.0)
fig = cg.beautify(fig)

fig.show()
```

```python
# Render a turntable GIF rotating around the z-axis
path = cg.animate(fig, "rotation.gif", axis="z", n_frames=100, duration=60)
Image(filename=path)
```
