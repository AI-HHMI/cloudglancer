```python
import numpy as np
import cloudglancer as cg
```

```python
# Create multiple point clouds
points1 = np.random.randn(200, 3)
points2 = np.random.randn(200, 3) * 2

# Create individual plots
fig1 = cg.plot(points1, title="Dataset 1")
fig2 = cg.plot(points2, title="Dataset 2")
```

```python
# Combine into a single figure
combined = cg.combine_plots([fig1, fig2], rows=1, cols=2)
combined.show()
```
