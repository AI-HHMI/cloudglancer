```python
import cloudglancer as cg
import numpy as np
```

```python
batch = np.random.randn(16, 500, 3)  # (B, N, 3)
```

```python
fig = cg.plot_grid(batch, colors="#1f77b4", size=1.0)
fig = cg.beautify(fig)
fig.show()
```

```python
# Combine with animate() for a rotating GIF of the whole grid
cg.animate(fig, "grid.gif", n_frames=60, width=1000, height=1000)
```
